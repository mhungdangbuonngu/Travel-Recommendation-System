```python
import psycopg2
import time
```

```python
import getpass
import os

postgres_url = getpass.getpass("Enter your postgresql url: ")
```

```python
def get_info():
    print(f'getting info for rooms dont have name')
    conn=psycopg2.connect(postgres_url)
    cur=conn.cursor()
    cur.execute("""SET search_path TO travel_database, public;""")
    query="""SELECT hprice_id, capacity, price from hotelprice WHERE room_type = 'N/a'  """
    cur.execute(query,)
    conn.commit()
    a=cur.fetchall()
    cur.close()
    conn.close()
    return a
    
```

```python
def generate_name(capacity,price,thresholds):
    if price < thresholds['low']:
        room_desc='standard'
    if price < thresholds['medium']:
        room_desc='middle class'
    else:
        room_desc='high-class'
    if capacity==1:
        cap='single room'
    elif capacity==2:
        cap='double room'
    elif capacity==3:
        cap='triple room'
    else:
        cap='family room'
    
    room_name=f"{room_desc} {cap}"

    return room_name
```

```python
def insert_name(id,name):
    print(f"update name for hprice_id: {id}")
    conn=psycopg2.connect(postgres_url)
    cur=conn.cursor()
    cur.execute("""SET search_path TO travel_database, public;""")
    query="""UPDATE hotelprice Set room_type = %s Where hprice_id = %s"""
    cur.execute(query,(name,id))
    conn.commit()
    cur.close()
    conn.close()
    print(f'inserted name for hprice_id:{id}')
    
```

```python
if __name__=='__main__':
    rooms=get_info()
    print(f'there are {len(rooms)} rooms that dont have price!!!')
    thresholds={'low':500000,'medium':1000000}
    for room in rooms:
        hprice_id,capacity,price=room
        room_name=generate_name(capacity,price,thresholds)
        insert_name(hprice_id,room_name)

```
